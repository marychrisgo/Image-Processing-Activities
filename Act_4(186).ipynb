```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
from scipy import ndimage
from skimage import feature, filters
import cv2
from PIL import Image
```

```python
#theoretical = 12.11 cm^2
square = mpimg.imread('square.png')
squareplot = plt.imshow(square)
plt.show()
```

```python
edge_horizont = ndimage.sobel(square, 0)
edge_vertical = ndimage.sobel(square, 1)
mag_square = np.hypot(edge_horizont, edge_vertical)
plt.imshow(mag_square)
```

```python
img = cv2.imread('square_edge.png', 0)
edges = cv2.Canny(img, 100, 255)  
indices = np.where(edges != [0])
coordinates = zip(indices[0], indices[1])
print(indices)
#print(coordinates)
```
